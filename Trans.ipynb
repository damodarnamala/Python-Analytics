```python
import numpy as np
import pandas as pd
import xlrd
df = pd.read_excel('data.xls')
df['Rate'] = df['Price'] * df['Qty']
xs = df[['Stock', 'Buy/Sell','Status','Qty','Rate']]
xs
```

```python
rslt_df = xs[xs['Status'] == 'Executed'] 
rslt_df
```

```python
final_df = rslt_df.sort_values(by=['Stock'], ascending=True)
final_df
```

```python
final_df = final_df.rename(columns={'Buy/Sell': 'BS'})
final_df
```

```python
final_df.to_csv('list.csv')
```
